```python
import random

''''
Esta función crea un tablero de juego de 10x10, 10 filas y 10 columnas. Cada casilla representada
por un _.
'''

def crear_tablero():
    '''''
    definimos el tamaño del tablero que será de 10 y creamos un tablero como una lista vacía
    '''
    tamaño = 10
    tablero = []
    '''  
    creamos cada fila del tablero con el bucle for. Range genera una secuencia de numeros
    de 0 hasta tamaño -1. por lo que el bucle se va a ejecutar 10 veces, una por cada fila
    e irá añadiendo el guión bajo.
    '''
    for _ in range(tamaño):  
        fila = ["_"] * tamaño
        tablero.append(fila)
    return tablero

'''  
se genera un barco aleatorio en un tablero:
-eslora=longitud del barco
-tamaño=tamaño del tablero(numero de filas y numero de columnas)
la función nos tiene que dar:
-horientación del barco(horizontal o vertical)
-coordenadas:que serán listas de tuplas con las casilas que ocupa el barco

'''
def crear_barco(eslora, tamaño):
    direccion = random.choice(["horizontal", "vertical"])
    if direccion == "horizontal":
        fila = random.randint(0, tamaño - 1)
        '''  
        selecciona una fila fija aleatoria para el barco, un numero entre el 0 y el 9
        '''
        columna_inicio = random.randint(0, tamaño - eslora)
        coordenadas = [(fila, columna_inicio + i) for i in range(eslora)]
        '''  
        Selecciona la posición inicial de la columna para que el barco no se salga del tablero.
        Si el barco tiene una eslora de 3 y el tamaño del tablero es 10, 
        la columna inicial será un número entre 0 y 7 (así hay espacio para las 3 casillas).
        y se genera una lista con las coordenadas del barco
        '''
    else:
        '''  
        si es vertical, se fijará una columna,se selecciona la posicion inicial de la fila 
        para que el barco no se salga del tablero y se genera la lista de coordenadas
        '''
        columna = random.randint(0, tamaño - 1)
        fila_inicio = random.randint(0, tamaño - eslora)
        coordenadas = [(fila_inicio + i, columna) for i in range(eslora)]
    return direccion, coordenadas

'''
se coloca la lista de barcos en el tablero:
se crea lista barcos_eslora, con las longitudes de los barcos y nos devolverá
una lista de tuplas con la direccion y coordenadas de cada barco
'''
def colocar_barco_en_tablero(tablero):
    barcos_eslora = [2, 2, 2, 3, 3, 4]
    lista_barcos = []

    '''  
    Para cada barco en la lista de eslora, se intentan colocar hasta que haya una posicion válida
    '''
    for eslora in barcos_eslora:
        while True:
            direccion, barco = crear_barco(eslora, len(tablero))
            if all(tablero[fila][columna] == "_" for fila, columna in barco):
                for fila, columna in barco:
                    tablero[fila][columna] = "O"
                lista_barcos.append((direccion, barco))
                break
    return lista_barcos

'''
se muestra el tablero del Usuario con barcos visibles. Recorre cada fila del tablero
y une los elementos de la fila con los espacios para mostrarlo como string
'''
def mostrar_tablero_usuario(tablero):
    print("\nTablero del Usuario:")
    for fila in tablero:
        print(" ".join(fila))

'''
se muestra el tablero del PC sin revelar los barcos, solo se muestran los disparos acertados o agua
'''
def mostrar_tablero_pc(tablero):
    print("\nTablero del PC:")
    '''  
    Se recorre cada fila del tablero, donde X se mantiene como X, A se mantiene como Agua
    y cualquier casilla con guion bajo o "O" se mantine como guión bajo.
    '''
    for fila in tablero:
        print(" ".join("X" if casilla == "X" else "A" if casilla == "A" else "_" for casilla in fila))


def disparar(casilla, tablero, barcos):
    ''' 
    Marca"X" si se acierta en una parte del barco
    Marca "A" si el disparo cae en agua
    Lleva un control de los barcos hundidos

    Los argumentos de la funcion son casillas(tupla),tablero(lista),barcos(lista)


    '''
    fila, columna = casilla
    ''' 
    se crea una lista de barcos_hundidos que almacena los barcos que ya han sido 
    procesados como undidos y se actualiza cada vez que un barco cumple la condición de estar
    completamente hundidos
    '''
    barcos_hundidos=[]
    if tablero[fila][columna] == "O":# Si hay una parte de un barco
        print(f"Disparo en {casilla}: ¡Tocado!")
        tablero[fila][columna] = "X" # Marca como tocado en el tablero con una X
        '''  
        se comprueba que el barco esta totalmente tocado, si todas las partes están tocadas
        '''
        for direccion, barco in barcos:
            if all(tablero[fila][columna] == "X" for fila, columna in barco):
                print(f"¡Barco hundido! Coordenadas: {barco}")
                barcos_hundidos.append(barco)
           
    elif tablero[fila][columna] == "_":#si la casilla está vacía es agua
        print(f"Disparo en {casilla}: ¡Agua!")
        tablero[fila][columna] = "A"
    else: #si ya se disparó en una casilla previamente
        print(f"Disparo en {casilla}: Ya disparaste aquí.")

'''
se verifica cada barco, se recorren todos los barcos para comprobar si alguno ha sido hundido
devuelve true si no queda ningun barco y false si aún hay barcos
'''
def todos_hundidos(tablero):
    return not any("O" in fila for fila in tablero) 

''''
Se gestionan los turnos del juego entre el usuario y el pc, alterna los turnos
hasta que los barcos de un jugador estén hundidos
'''
def turnos(tablero_usuario, tablero_pc, barcos_usuario, barcos_pc):
    turno = "usuario" #El usuario empieza disparando
    while True:
        if turno == "usuario": #este es el turno del usuario
            print("\nTu turno:")
            mostrar_tablero_usuario(tablero_usuario)
            mostrar_tablero_pc(tablero_pc)
            while True:#para el solicitar el disparo en la fila
                try:
                    fila = int(input("Introduce la fila del disparo (0-9): "))
                    if 0<=fila<=9:
                        break #sale del bucle si la tirada es válida
                    else:
                        print("Por favor. Introduce un número entre el 0 y el 9")
                except ValueError:
                    print("Entrada no válida. Introduce un número válido")
                
            while True:#para solicitar el disparo en la coulmna
                try:
                    columna = int(input("Introduce la columna del disparo (0-9): "))
                    if 0<=columna<=9:
                        break
                    else:
                        print("Por favor. Introduce un número entre el 0 y el 9")
                except ValueError:
                    print("Entrada no válida. Introduce un número válido")

            disparar((fila, columna), tablero_pc, barcos_pc)
            '''  
            llamada a la funcion disparar, los argumentos son las coordenadas del disparo 
            del usuario, el tablero del pc donde se realizan los disparos y la lista de
            barcos del pc donde se verifica los tocados o hundidos.

            La fundicon todos hundidos recorre el tablero buscando si quedan barcos "O" sin hundir
            si no hay barcos debuelve True, por lo que sale el mensaje de victoria y sale del bucle
            '''
            
            if todos_hundidos(tablero_pc):
                print("¡Felicidades! Hundiste todos los barcos del PC. ¡Ganaste!")
                break
            '''  
            el cambio de turno se da cuando comprueba que la casilla comparada no contiene X
            si es así, será el turno del pc
            '''
            if tablero_pc[fila][columna] != "X":
                turno = "pc"
        else:
            print("\nTurno del PC:")
            fila = random.randint(0, 9)#el pc dispara da coordenads de disparo de forma aleatoria
            columna = random.randint(0, 9)
            print(f"El PC dispara en ({fila}, {columna})")
            disparar((fila, columna), tablero_usuario, barcos_usuario)
            mostrar_tablero_usuario(tablero_usuario)
            '''  
            comprueba que el usuario tiene barcos restantes
            '''
            if todos_hundidos(tablero_usuario):
                print("¡Lo siento! El PC hundió todos tus barcos. ¡Perdiste!")
                break
            if tablero_usuario[fila][columna] != "X":
                turno = "usuario"

# Flujo principal
tablero_usuario = crear_tablero()
tablero_pc = crear_tablero()
barcos_usuario = colocar_barco_en_tablero(tablero_usuario)
barcos_pc = colocar_barco_en_tablero(tablero_pc)
turnos(tablero_usuario, tablero_pc, barcos_usuario, barcos_pc)
```
